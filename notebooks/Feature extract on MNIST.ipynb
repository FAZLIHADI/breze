```python
def scale_cols_to_unit_interval(arr):
    arr = arr.copy()
    arr -= arr.min(axis=0)
    arr /= arr.max(axis=0)
    arr *= 255
    return arr.astype('uint8')
    

def save_and_display(arr, fname):
    pilimg = Image.fromarray(arr)
    pilimg.save(fname)
    return image(filename=fname, width=600)
```

Prepare data
============

```python
datafile = '../examples/mnist.pkl.gz'
# Load data.                                                                                                   

with gzip.open(datafile,'rb') as f:                                                                        
    train_set, val_set, test_set = cPickle.load(f)                                                       

X, Z = train_set                                                                                               
VX, VZ = val_set
TX, TZ = test_set

image_dims = 28, 28
```

Common Hyperparameters
======================

```python
n_features = 128
max_passes = 100
```

Feature Learning
================

What follows is the code to make a couple of feature extractors run. 
The idea is to only run a single FE, which will then instantiate several
local variables which can be inspected for comparis. That is

  - An array of size ``(i, f)`` called ``filters``. ``i`` is the dimensionality
    of the input, while ``f`` is the number of features. The matrix can be used
    to do (linear) feature extraction by multiplying it from the right to an 
    input vector.
  - Arrays ``F``, ``VF`` and ``TF`` which are the features extracted from the training,
    validation and test set respectively. Each is an array of the size ``(n, f)`` where
    ``n`` is the amount of samples in the respective set.

Sparse Filtering
----------------

```python
from brummlearn.sparsefiltering import SparseFiltering

max_iter = max_passes

sf = SparseFiltering(X.shape[1], n_features)
f_loss = sf.function(['inpt'], 'loss')


stop = climin.stops.any_([
    climin.stops.converged('loss'),
    climin.stops.rising('val_loss', 5, 1e-5, patience=5),
    climin.stops.after_n_iterations(max_iter),
    ])

train_losses = []
val_losses = []

for i, info in enumerate(sf.iter_fit(X)):
    info['val_loss'] = f_loss(VX)
    train_losses.append(info['loss'])
    val_losses.append(info['val_loss'])
    
    print train_losses[-1], val_losses[-1]
    
    if stop(info):
        break  
        
filters = sf.parameters['in_to_feature']
F = sf.transform(X)
VF = sf.transform(VX)
TF = sf.transform(TX)
```

Classification Performance with Logistic Regression
===================================================

```python
best_loss = float('inf')
best = None
for c in [0.01, 0.1, 1., 2.]:
    clf = linear_model.LogisticRegression(C=c)
    clf.fit(F, Z)
    loss = (clf.predict(VF) != VZ).mean()                  
    if loss < best_loss:
        best = clf
        best_loss = loss
    print 'c, loss', c, loss 
    
print '#wrong', 10000 - (clf.predict(TF) == TZ).sum()
```
