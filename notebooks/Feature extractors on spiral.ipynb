```python
import numpy as np
import brummlearn.autoencoder as ae
from brummlearn import rica
from brummlearn import sparsefiltering
import climin.stops
import brummlearn.data
```

```python
def spiral(angle, a=1.0, b=1.0):
    radius = a * np.exp(b * angle)
    x = radius * np.sin(angle)
    y = radius * np.cos(angle)
    return np.concatenate([x[:, np.newaxis], y[:, np.newaxis]], axis=1)
```

```python
X = spiral(np.linspace(0.001, 10, 1000), a=1, b=.2)
#X = np.log(abs(X) + 1)
#x = np.linspace(0, 1, 50)[:, np.newaxis]
#X = np.concatenate([x**2, x], axis=1)
X -= X.mean(axis=0)
X /= X.std(axis=0)
X += np.random.normal(0, 0.01, X.shape)
fig = plt.figure(figsize=(16, 9))
ax = fig.add_subplot(111)
ax.plot(X[:, 0], X[:, 1], 'x')
brummlearn.data.shuffle(X)
```

```python
opt = 'rmsprop', {'steprate': 1e-4, 'momentum': 0.95, 'decay': 0.9, 'step_adapt': 0.05}
batch_size = 50
max_passes = 100
batches_per_pass = X.shape[0] / batch_size
max_iter = batches_per_pass * max_passes
n_hidden = 10
method = 'dae'
if method == 'dae':
    model = ae.DenoisingAutoEncoder(2, n_hidden, batch_size=batch_size, optimizer=opt, c_noise=0.2, hidden_transfer='softplus')
    f_energy = lambda x: ((x - model.reconstruct(x))**2).sum()
elif method == 'cae':
    model = ae.ContractiveAutoEncoder(2, n_hidden, batch_size=batch_size, optimizer=opt, c_jacobian=1000, hidden_transfer='tanh', reconstruct_loss='squared')
    f_energy = lambda x: ((x - model.reconstruct(x))**2).sum()
elif method == 'sae':
    model = ae.SparseAutoEncoder(2, n_hidden, batch_size=batch_size, optimizer=opt, c_sparsity=1, hidden_transfer='sigmoid', reconstruct_loss='squared')
    f_energy = lambda x: ((x - model.reconstruct(x))**2).sum()
elif method == 'rica':
    model = rica.Rica(2, n_hidden, c_ica=10, optimizer=opt, hidden_transfer='sigmoid')
    f_energy = lambda x: model.transform(x).sum()
elif method == 'sf':
    model = sparsefiltering.SparseFiltering(2, n_hidden, optimizer=opt)
    f_energy = lambda x: model.transform(x).sum()   
```

```python
stop = climin.stops.after_n_iterations(max_iter)
report = climin.stops.modulo_n_iterations(batches_per_pass * 10)
for i, info in enumerate(model.powerfit((X,), (X,), stop=stop, report=report)):
    print info['loss']
```
