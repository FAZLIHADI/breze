Sparse Coding
=============

This notebook demonstrates Sparse Coding on the natural image patches by
Aapo Hyvärinen.

```python
import time

import h5py
from PIL import Image
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import numpy as np
import theano.tensor as T

import climin.stops
import climin.initialize

from breze.learn.sparsecoding import SparseCoding
from breze.learn.utils import tile_raster_images
import breze.learn.base

from IPython.core import pylabtools
pylabtools.figsize(16, 9)
from IPython.core.display import Image as image
```

### Helper functions
This will help us later with plotting.

```python
def scale_cols_to_unit_interval(arr):
    arr = arr.copy()
    arr -= arr.min(axis=0)
    arr /= arr.max(axis=0)
    arr *= 255
    return arr.astype('uint8')
```

Prepare data
============

The data can be downloaded as part of Quoc Le's rica code here: http://cs.stanford.edu/~quocle/rica_release.zip.

```python
datafile = 'hyv_patches_16.mat'
# Load data.                                                                                                   

f = h5py.File(datafile)
patches = np.array(f['patches'])

X = patches[:20000]
VX = patches[20000:30000]
TX = patches[30000:]

image_dims = 16, 16

X, VX, TX = [breze.learn.base.cast_array_to_local_type(i) for i in X, VX, TX]
```

```python
img = tile_raster_images(X[:1024], image_dims, (32, 32), (1, 1))
# We write the filters to a file, since I do not like the matplot imshow functionality. :)
imshow(img, cmap=cm.binary)
```

```python
n_features = 32
c_sparsity = 5

max_passes = 50
batch_size = 2000
max_iter = max_passes * X.shape[0] / batch_size
n_report = X.shape[0] / batch_size

fe = SparseCoding(X.shape[1], n_features, c_sparsity=c_sparsity, batch_size=batch_size)
losses = []
```

```python
print "max_iter", max_iter

stop = climin.stops.after_n_iterations(max_iter)
pause = climin.stops.modulo_n_iterations(n_report)
start = time.time()

header = '#\tseconds\tloss'
print header
print '-' * len(header)
    
for i, info in enumerate(fe.iter_fit(X)):
    if pause(info):
        info['time_passed'] = time.time() - start
        losses.append(info['loss'])
        row = '%(n_iter)g\t%(time_passed)g\t%(loss)g' % info
        print row
    if stop(info):
        break
```

```python
F = fe.transform(X)
```

```python
R = fe.inverse_transform(F)
```

```python
img = tile_raster_images(R[:1024], image_dims, (32, 32), (1, 1))
# We write the filters to a file, since I do not like the matplot imshow functionality. :)
plt.imshow(img, cmap=cm.binary)
```

```python
img = tile_raster_images((X[:1024] - R[:1024]), image_dims, (32, 32), (1, 1))
# We write the filters to a file, since I do not like the matplot imshow functionality. :)
plt.imshow(img, cmap=cm.binary)
```
