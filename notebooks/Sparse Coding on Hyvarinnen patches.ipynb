Sparse Coding
=============

This notebook demonstrates Sparse Coding on the natural image patches by
Aapo Hyvärinen.

```python
import h5py
from PIL import Image
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import numpy as np
import theano.tensor as T

import climin.stops
import climin.initialize

from brummlearn.sparsecoding import SparseCoding
from brummlearn.utils import tile_raster_images

from IPython.core import pylabtools
pylabtools.figsize(16, 9)
from IPython.core.display import Image as image
```

Helper functions
================

```python
def scale_cols_to_unit_interval(arr):
    arr = arr.copy()
    arr -= arr.min(axis=0)
    arr /= arr.max(axis=0)
    arr *= 255
    return arr.astype('uint8')
    

def save_and_display(arr, fname):
    pilimg = Image.fromarray(arr)
    pilimg.save(fname)
    return image(filename=fname, width=600)
```

Prepare data
============

```python
datafile = 'hyv_patches_16.mat'
# Load data.                                                                                                   

f = h5py.File(datafile)
patches = np.array(f['patches'])

X = patches[:20000]
VX = patches[20000:30000]
TX = patches[30000:]

image_dims = 16, 16
```

```python
img = tile_raster_images(X[:1024], image_dims, (32, 32), (1, 1))
# We write the filters to a file, since I do not like the matplot imshow functionality. :)
save_and_display(img, 'example_patches.png')
```
