TSNE applied to the IRIS data set
=================================

We will apply TSNE to the IRIS data set in this notebook. The data can be downloaded from here: http://archive.ics.uci.edu/ml/datasets/Iris.

Start out with some imports.

```python
import numpy as np

from IPython.core import pylabtools
pylabtools.figsize(16, 9)

from brummlearn.tsne import Tsne
```

Prepare data.

```python
data = open('iris.data').readlines()
X = np.empty((len(data), 4))
rows = (i.split(',') for i in data if i.strip())
rows = (([float(j) for j in i[:-1]], i[-1]) for i in rows)
classes = []
for i, (inpt, klass) in enumerate(rows):
    X[i] = inpt
    classes.append(klass.strip())
    
class_types = sorted(set(classes))
class_to_int = dict(zip(class_types, range(3)))
classes = np.array([class_to_int[i] for i in classes])
```

Instantiate Tsne object and run the optimization.

```python
tsne = Tsne(4, 2, max_iter=1000, verbose=True)
E = tsne.fit_transform(X)
```

Visualize the found embeddings.

```python
pylab.plot(E[classes == 0, 0], E[classes == 0, 1], 'ro', label=class_types[0])
pylab.plot(E[classes == 1, 0], E[classes == 1, 1], 'yo', label=class_types[1])
pylab.plot(E[classes == 2, 0], E[classes == 2, 1], 'ko', label=class_types[2])
pylab.legend()
```
