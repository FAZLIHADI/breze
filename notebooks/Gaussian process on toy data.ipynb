```python
import numpy as np

from breze.learn.gaussianprocess import GaussianProcess
import breze.learn.base
```

Make Data
=========

We will generate a data set with homoscedastic noise. GPs are not well suited for that, but the man will be learned ok anyways.

```python
X = np.linspace(0, 1, 100)[:, np.newaxis]
Z = (X - 0.5)**3 + np.random.normal(0, 0.01, (X.shape[0], 1))
Z[np.where(X > 0.5)] += np.random.normal(0, 0.1, np.where(X > 0.5)[0].size)

X, Z = [breze.learn.base.cast_array_to_local_type(i) for i in (X, Z)]
```

Let's have a look at the data.

```python
fig = plt.figure(figsize=(16, 9))
ax = fig.add_subplot(111)
ax.plot(X[:, 0], Z[:, 0], '+')
```

Learning
========

```python
model = GaussianProcess(1, kernel='ardse', optimizer='lbfgs', max_iter=100)
model.fit(X, Z)
```

Evaluation
==========

```python
ticks = breze.learn.base.cast_array_to_local_type(np.linspace(0, 1, 100)[:, np.newaxis])
Y, S = model.predict(ticks, var=True)
```

```python
fig = plt.figure(figsize=(16, 9))
ax = fig.add_subplot(111)
ax.plot(ticks[:, 0], Y[:, 0], '-')
ax.plot(X[:, 0], Z[:, 0], 'r+', markersize=20)
ax.fill_between(ticks[:, 0], Y[:, 0] - S[:, 0], Y[:, 0] + S[:, 0], alpha=.2)
```
