```python
import numpy as np
from brummlearn.gaussianprocess import GaussianProcess
```

```python
X = np.linspace(0, 1, 100)[:, np.newaxis]
Z = (X - 0.5)**3 + np.random.normal(0, 0.01, (X.shape[0], 1))
Z[np.where(X > 0.5)] += np.random.normal(0, 0.1, np.where(X > 0.5)[0].size)
```

```python
fig = plt.figure(figsize=(16, 9))
ax = fig.add_subplot(111)
ax.plot(X[:, 0], Z[:, 0], '+')
```

```python
model = GaussianProcess(1, kernel='rbf', optimizer='lbfgs', max_iter=100)
model.fit(X, Z)
```

```python
ticks = np.linspace(0, 1, 100)[:, np.newaxis]
Y, S = model.predict(ticks, var=True)
```
