ICA with reconstruction cost
============================

This notebook demonstrates ICA with reconstruction cost on the natural image patches by
Aapo Hyvärinen.

```python
import h5py
from PIL import Image
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import numpy as np
import theano.tensor as T

import climin.stops
import climin.initialize

from sklearn import linear_model

from brummlearn.pca import pca
from brummlearn.utils import tile_raster_images

from IPython.core import pylabtools
pylabtools.figsize(16, 9)
from IPython.core.display import Image as image
```

Helper functions
================

```python
def scale_cols_to_unit_interval(arr):
    arr = arr.copy()
    arr -= arr.min(axis=0)
    arr /= arr.max(axis=0)
    arr *= 255
    return arr.astype('uint8')
    

def save_and_display(arr, fname):
    pilimg = Image.fromarray(arr)
    pilimg.save(fname)
    return image(filename=fname, width=600)

def fit(X, VX, fe, stop, pause):
    train_losses = []
    val_losses = []
    f_loss = fe.function(['inpt'], 'loss')
    
    best_pars = None
    best_loss = float('inf')

    for info in fe.iter_fit(X):
        if not pause(info):
            continue
        info['val_loss'] = f_loss(VX)
        train_losses.append(info['loss'])
        val_losses.append(info['val_loss'])
        
        if info['val_loss'] < best_loss:
            best_loss = info['val_loss']
            best_pars = fe.parameters.data.copy()
            
        info['best_loss'] = best_loss
        info['best_pars'] = best_pars
        
        yield info
        if stop(info):
            print 'stopping condition'
            break
```

Prepare data
============

```python
datafile = 'hyv_patches_16.mat'
# Load data.                                                                                                   

f = h5py.File(datafile)
patches = np.array(f['patches'])

#patches -= patches.mean(axis=1)[:, np.newaxis]
#patches /= patches.std(axis=1)[:, np.newaxis]

X = patches[:20000]
VX = patches[20000:30000]
TX = patches[30000:]

image_dims = 16, 16
```

```python
img = tile_raster_images(X[:1024], image_dims, (32, 32), (1, 1))
# We write the filters to a file, since I do not like the matplot imshow functionality. :)
save_and_display(img, 'example_patches.png')
```

Common Hyperparameters
======================

```python
n_features = 64
max_passes = 300
```

ICA with Reconstruction Cost
============================

```python
losses, V_losses = [], []
rec_losses, ica_losses = [], []

from brummlearn.rica import Rica

c_ica = 0.1
batch_size = 20000
hidden_transfer = 'identity'
feature_transfer = 'softabs'
out_transfer = 'identity'
loss = 'squared'
optimizer = 'lbfgs', {'n_factors': 50}

fe = Rica(X.shape[1], n_features, c_ica=c_ica, 
          hidden_transfer=hidden_transfer, feature_transfer=feature_transfer, out_transfer=out_transfer,
          loss=loss, optimizer=optimizer)
fe.parameters.data[...] = np.random.normal(0, 1e-2, fe.parameters.data.shape)
climin.initialize.sparsify_columns(fe.parameters['in_to_hidden'], 1)
```

```python
batches_per_pass = 20000. / batch_size
max_iter = int(math.ceil(max_passes * batches_per_pass))

print "max_iter", max_iter

stop = climin.stops.any_([
    #climin.stops.converged('loss'),
    #climin.stops.rising('val_loss', 10, 1e-5, patience=5),
    climin.stops.after_n_iterations(max_iter),
    ])

pause = climin.stops.modulo_n_iterations(10)

# In order to inspect some statistics during learning, we collect some 
# statistics:
#  - the losses of the reconstruction and the ica part of the objective,
#  - the ratio of the length of their gradients,
#  - TODO: the lengths of their gradients.

def length(x): return T.sqrt((x**2).sum())
l_reconstr_grad = length(T.grad(fe.exprs['reconstruct_loss'], fe.parameters.flat))
l_ica_grad = length(T.grad(fe.exprs['ica_loss'], fe.parameters.flat))
f_statistics = fe.function(['inpt'], ['reconstruct_loss', l_reconstr_grad, 'ica_loss', l_ica_grad])

keys = '#', 'loss', 'val loss', 'step_length', 'rec loss', 'ica loss', '|dr|/|di|', '|dr|', '|di|'
max_len = max(len(i) for i in keys)
header = ''.join(i.ljust(max_len + 2) for i in keys)
print header
print '-' * len(header)
    
for i, info in enumerate(fit(X, VX, fe, stop, pause)):
    losses.append(info['loss'])
    V_losses.append(info['val_loss'])
    rec_loss, rec_grad, ica_loss, ica_grad = f_statistics(X)
    rec_losses.append((rec_loss, rec_grad))
    ica_losses.append((ica_loss, ica_grad))
    
    img = tile_raster_images(fe.parameters['in_to_hidden'].T, image_dims, (8, 8), (1, 1))
    # We write the filters to a file, since I do not like the matplot imshow functionality. :)
    save_and_display(img, 'filters-%i.png' % i)  
    
    row = ('%i' % info['n_iter'], '%.4f' % info['loss'], '%.4f' % info['val_loss'],
           '%.4f' % info['step_length'], '%.4f' % rec_loss, '%.4f' % ica_loss,
           '%.4f' % (rec_grad / ica_grad), '%.4f' % rec_grad, '%.4f' % ica_grad)
    
    
    print ''.join(i.ljust(max_len + 2) for i in row)
```

```python
filters = fe.function([], 'in_to_hidden_normed')()
#filters = np.dot(w, filters)
F = fe.transform(X)
VF = fe.transform(VX)
TF = fe.transform(TX)

plots = [
    {'training error': losses,
     'validation error': V_losses
     },
    {'reconstruction loss' :[i for i, _ in rec_losses],
     'ica loss': [i * c_ica for i, _ in ica_losses]
     },
    {'ratio of rec/ica losses': [i / (j * c_ica) for (i, _), (j, _) in zip(rec_losses, ica_losses)]},
]   
```

```python
f_reconstruct = fe.function(['inpt'], 'output')
TR = f_reconstruct(TX[:64])

img = tile_raster_images(TR, image_dims, (8, 8), (1, 1))
# We write the filters to a file, since I do not like the matplot imshow functionality. :)
save_and_display(img, 'reconstructions.png')
```

```python
img = tile_raster_images(TR - TX[:64], image_dims, (8, 8), (1, 1))
# We write the filters to a file, since I do not like the matplot imshow functionality. :)
save_and_display(img, 'reconstruction-errors.png')
```
