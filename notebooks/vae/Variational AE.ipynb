```python
datafile = '../mnist.pkl.gz'
# Load data.                                                                                                   

with gzip.open(datafile,'rb') as f:                                                                        
    train_set, val_set, test_set = cPickle.load(f)                                                       

X, Z = train_set                                                                                               
VX, VZ = val_set
TX, TZ = test_set

Z = one_hot(Z, 10)
VZ = one_hot(VZ, 10)
TZ = one_hot(TZ, 10)

image_dims = 28, 28

X, Z, VX, VZ, TX, TZ = [cast_array_to_local_type(i) for i in (X, Z, VX,VZ, TX, TZ)]
```

```python
reload(vae)
batch_size = 200
optimizer = 'rmsprop', {'step_rate': 1e-4, 'momentum': 0.95, 'decay': .95, 'offset': 1e-6}
#optimizer = 'adadelta', {'step_rate': 1, 'momentum': 0.0, 'decay': .95, 'offset': 1e-6}
m = vae.VariationalAutoEncoder(X.shape[1], [200, 200, 200], [200, 200], ['rectifier'] * 2, ['rectifier'] * 2, 
                               latent_prior='white_gaussian', 
                               latent_posterior='diag_gaussian',
                               visible='bernoulli',
                               optimizer=optimizer, batch_size=batch_size)

#m.exprs['loss'] += 0.001 * (m.parameters.enc_in_to_hidden ** 2).sum() / m.exprs['inpt'].shape[0]

climin.initialize.randomize_normal(m.parameters.data, 0, 1e-3)

#climin.initialize.sparsify_columns(m.parameters['enc_in_to_hidden'], 15)
#climin.initialize.sparsify_columns(m.parameters['enc_hidden_to_hidden_0'], 15)
#climin.initialize.sparsify_columns(m.parameters['dec_hidden_to_out'], 15)

f_latent_mean = m.function(['inpt'], 'latent_mean')
f_latent_mean_var = m.function(['inpt'], 'latent_mean')
f_sample = m.function(['sample'], 'dec_output')
f_recons = m.function(['inpt'], 'dec_output')
f_transform = m.function(['inpt'], 'enc_output')
```

### Denoising

```python
def saltnpepper(X, p=0.1):
    to_pepper = np.random.random(X.shape) < p
    to_salt = np.random.random(X.shape) < p
    X = X.copy()
    X[to_pepper] = 1
    X[to_salt] = 0
    return X

def denoise(m, X):
    H = f_latent_mean(X)
    return f_sample(H)
```

```python
noised = saltnpepper(X[:64], 0.05)
denoised = denoise(m, noised)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(18, 9))

img = tile_raster_images(noised, image_dims, (8, 8), (1, 1))
axs[0].imshow(img, cmap=cm.binary)

img = tile_raster_images(denoised, image_dims, (8, 8), (1, 1))
axs[1].imshow(img, cmap=cm.binary)
```

```python
from breze.arch.component.distributions.mvn import logpdf
import theano.tensor as T
import theano
```

```python
sample = T.matrix()
mean = T.vector()
cov = T.matrix()
mvnlogpdf = logpdf(sample, mean, cov)
f_mvnlogpdf = theano.function([sample, mean, cov], mvnlogpdf)
```

```python
f_latent_mean_var = m.function(['inpt'], ['latent_mean', 'latent_var'])
f_rec_loss_of_sample = m.function(['sample', 'inpt'], 'rec_loss_sample_wise')

def marginal_ll(m, x, n_samples=10):
    mean, var = f_latent_mean_var(x[np.newaxis])
    
    s = np.random.standard_normal((n_samples, mean.shape[1]))
    #print 'x', s.sum()
    s *= np.sqrt(var)
    s += mean
    s = s.astype('float32')
    
    targets = np.concatenate([x[np.newaxis]] * n_samples)
   
    #print f_rec_loss_of_sample(s, targets)
    
    lp_v_given_s = -f_rec_loss_of_sample(s, targets)
    lq_s_given_v = f_mvnlogpdf(s, mean[0], np.diag(var.flatten()))
    lp_s = f_mvnlogpdf(s, np.zeros(s.shape[1]).astype('float32'), 
                   np.eye(s.shape[1]).astype('float32'))
    
    #print lp_v_given_s
    #print lq_s_given_v
    #print lp_s
    #print '--' * 20
                        
    ll = np.exp(lp_v_given_s + lp_s - lq_s_given_v).mean()
            
    return ll    

```

```python
ll = 0
for i in range(X.shape[0]):
    ll += np.log(marginal_ll(m, X[i], 2 ** 5))
    
ll /= 50000
print ll
```
