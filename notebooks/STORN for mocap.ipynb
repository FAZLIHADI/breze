```python
def preprocess_data(Motion):
 
    n_seq = Motion.shape[1]
 
    # assume data is MIT format for now
    indx = np.r_[
        np.arange(0,6),
        np.arange(6,9),
        13,
        np.arange(18,21),
        25,
        np.arange(30,33),
        37,
        np.arange(42,45),
        49,
        np.arange(54,57),
        np.arange(60,63),
        np.arange(66,69),
        np.arange(72,75),
        np.arange(78,81),
        np.arange(84,87),
        np.arange(90,93),
        np.arange(96,99),
        np.arange(102,105)]
 
    row1 = Motion[0,0][0]
 
    offsets =   np.r_[
        row1[None,9:12],
        row1[None,15:18],
        row1[None,21:24],
        row1[None,27:30],
        row1[None,33:36],
        row1[None,39:42],
        row1[None,45:48],
        row1[None,51:54],
        row1[None,57:60],
        row1[None,63:66],
        row1[None,69:72],
        row1[None,75:78],
        row1[None,81:84],
        row1[None,87:90],
        row1[None,93:96],
        row1[None,99:102],
        row1[None,105:108]]
 
    # collapse sequences
    batchdata = np.concatenate([m[:, indx] for m in Motion.flat], axis=0)
 
    data_mean = batchdata.mean(axis=0)
    data_std = batchdata.std(axis=0)
 
    batchdata = (batchdata - data_mean) / data_std
 
    # get sequence lengths
    seqlen = [s.shape[0] for s in Motion.flat]
 
 
    return batchdata, seqlen, data_mean, data_std
 
def load_data_rather_raw(filename):
    # load data post preprocess1
    mat_dict = loadmat(filename)
    Motion = mat_dict['Motion']
 
    batchdata, seqlen, data_mean, data_std = preprocess_data(Motion)
 
    # put data into shared memory
    x = np.asarray(batchdata, dtype=theano.config.floatX)
 
    return x, seqlen, data_mean, data_std


def load_data(x, seqlen, T=200):
    xs = [x[seqlen[0] + seqlen[1]:]]
    vxs = [x[seqlen[0]:seqlen[0] + seqlen[1]]]
    tx = x[:seqlen[0]]
    
    x = split(xs, T)
    vx = split(vxs, T)
    
    X, M = padzeros(x, False, True)
    VX, VM = padzeros(vx, False, True)
    
    X, M, VX, VM = [interleave(i) for i in (X, M, VX, VM)]
    
    print M.shape
    
    M = M.mean(2)[:, :, np.newaxis]
    VM = VM.mean(2)[:, :, np.newaxis]

    print 'size of X', X.size
    print 'size of VX', VX.size

    X, VX, M, VM = [base.cast_array_to_local_type(i) for i in (X, VX, M, VM)]
    tx = base.cast_array_to_local_type(tx)

    print 'training shape', X.shape, M.shape
    print 'validation shape', VX.shape, VM.shape

    return X, M, VX, VM, tx

x, seqlen, mean, std = load_data_rather_raw('motion.mat')
X, M, VX, VM, tx = load_data(x, seqlen, 100)

print 'masking ratios: train %g val %g' %(M.mean(), VM.mean())
```

```python
sample_map = m.vae.latent[:, :, :n_latents]
gen_inpt = T.concatenate([sample_map, m.vae.condition], axis=2)

map_recons_exprs = theano.clone(m.exprs['output'], 
                                {m.vae.sample: sample_map})
f_map_recons = m.function(['inpt'], map_recons_exprs)
```

```python
def map_impute(start, to_impute_idxs, n_steps=10):
    denoised = start.copy()
    denoised[to_impute_idxs] = np.random.standard_normal(denoised[to_impute_idxs].shape)
    noisy = denoised.copy()
    
    for i in range(n_steps):
        reconstructed = f_map_recons(denoised)
        denoised[to_impute_idxs] = reconstructed[:, :, :49][to_impute_idxs]
    return noisy, denoised
```

```python
clean = VX[:, :1]
mask = np.zeros_like(clean).astype('uint8')
mask[30:40][...] = 1
to_impute_idxs = np.where(mask)
#noised[to_impute_idxs] = np.random.standard_normal(noised[to_impute_idxs].shape) * 1

noisy, denoised = map_impute(clean, to_impute_idxs, 100)
```
