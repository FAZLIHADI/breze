```python
T = 100

r = np.linspace(0, 10, T)
X1 = (np.sin(r) + 0.05)[:, np.newaxis]
X2 = (np.cos(r * 2 + 1) + .3 * r)[:, np.newaxis]

X = np.concatenate([X1, X2], axis=1)

xs = [X[10:] + np.random.normal(0, .05, X[10:].shape), X[:-10] + np.random.normal(0, .05, X[:-10].shape)]
```

```python
fig, axs = plt.subplots(1,2, figsize=(16, 9))
axs[0].plot(xs[0])
axs[1].plot(xs[1])
```
