```python
from breze.learn import kmeans, pca
from breze.learn.utils import tile_raster_images

from skdata import cifar10
from sklearn.preprocessing import StandardScaler
```

```python
# From: http://stackoverflow.com/questions/12201577/convert-rgb-image-to-grayscale-in-python

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.144])
```

```python
ds = cifar10.views.OfficialImageClassification()
X0 = ds.train.x.reshape(-1, 3)
X1 = rgb2gray(X0)
X1 = X1.reshape((50000, -1))
```

```python
examples = X1[:64]
tile_size = 8, 8
image_dims = 32, 32

img = tile_raster_images(examples, image_dims, tile_size, (1, 1))
# We write the filters to a file, since I do not like the matplot imshow functionality. :)
plt.imshow(img, cmap=cm.binary)
```

```python
m0 = StandardScaler()
m0.fit(X1)
X2 = m0.transform(X1)
```

```python
m1 = pca.Zca(min_eig_val=1e-)
m1.fit(X2)
X3 = m1.transform(X2)
```

```python
examples = X3[:64]
tile_size = 8, 8
image_dims = 32, 32

img = tile_raster_images(examples, image_dims, tile_size, (1, 1))
# We write the filters to a file, since I do not like the matplot imshow functionality. :)
plt.imshow(img, cmap=cm.binary)
```

```python
import climin.initialize
```

```python
import itertools
```

```python
class SparseGainShapeKMeans(kmeans.GainShapeKMeans):
    
    def prepare(self, n_inpt):
        super(SparseGainShapeKMeans, self).prepare(n_inpt)
        climin.initialize.sparsify_columns(self.dictionary, 5)
        super(SparseGainShapeKMeans, self).normalize_dict()
        
    def iter_fit(self, X):
        self.prepare(X.shape[1])
        for i in itertools.count():
            code = self.transform(X, activation='omp-1')
            self.dictionary += np.dot(X.T, code)

            # If a cluster does not get any samples, reset it to a sample from
            # the training set.
            empty_clusters, = np.where(code.sum(axis=0) == 0)
            for i in empty_clusters:
                idx = np.random.randint(X.shape[0])
                self.dictionary[:, i] = X[idx]
                climin.sparsify_columns(self.dictionary[:, i:i+1], 5)

            self.normalize_dict()
            yield {'n_iter': i}
```

```python
m = SparseGainShapeKMeans(64, max_iter=250)
m.fit(X3)
```

```python
tile_size = 8, 16
image_dims = 32, 32

fig = plt.figure(figsize=(16, 9))
ax = fig.add_subplot(111)
img = tile_raster_images(m.dictionary.T[:128], image_dims, tile_size, (1, 1))
# We write the filters to a file, since I do not like the matplot imshow functionality. :)
ax.imshow(img, cmap=cm.binary)
```
