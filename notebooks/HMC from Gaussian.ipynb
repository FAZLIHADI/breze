Sampling from a Bivariate Gaussian with Hybrid Monte Carlo
==========================================================

```python
import numpy as np
from breze.learn.sampling import hmc
import matplotlib.mlab as mlab
```

Specifying the Gaussian and the energy function
-----------------------------------------------

```python
cov = np.array([[1., 1.3], [1.3, 2]])
start = np.array([.2, .5])
precision = np.linalg.inv(cov)
f_energy = lambda x: 0.5 * np.dot(x.T, np.dot(precision, x))
f_d_energy_wrt_x = lambda x: np.dot(precision, x)
```

Generating a Mesh Grid for Visualization
----------------------------------------

```python
delta = .1
x = np.arange(-10.0, 10.0, delta)
y = np.arange(-10.0, 10.0, delta)
X, Y = np.meshgrid(x, y)
Z = mlab.bivariate_normal(X, Y, cov[0, 0], cov[1, 1], 0, 0, cov[0, 1])
```

Sampling
--------

```python
n_samples = 100
n_steps = 1000
samples = hmc.sample(f_energy, f_d_energy_wrt_x, start, n_steps, initial_step_size=0.5)
samples = np.concatenate([j[np.newaxis] for i, j in zip(range(n_samples), samples)] ,axis=0)
```
