```python
import mlexperiments.data.ucr as ucr

from brummlearn.rnn import SupervisedRnn
import brummlearn.data

from climin.initialize import sparsify_columns, bound_spectral_radius
from climin.stops import rising, converged, after_n_iterations, modulo_n_iterations, time_elapsed

from sklearn import neighbors
```

```python
dataset = ucr.TwoPatterns()
```

```python
dataset.pars = {'val-fraction': None}
dataset.prepare_data()
X, Z, TX, TZ = dataset.X, dataset.Z, dataset.TX, dataset.TZ
```

```python
offset, division = 'mean', 'std'

if offset == 'begin':
    TX -= TX[0, :, :]
    X -= X[0, :, :]
if offset == 'rowwise_mean':
    TX -= TX.mean(axis=1)[:, np.newaxis, :]
    X -= X.mean(axis=1)[:, np.newaxis, :]
if offset == 'mean':
    TX -= X.mean()
    X -= TX.mean()
if offset == 'to_zero':
    TX -= X.min()
    X -= X.min()
    
if division == 'rowwise_std':
    TX /= TX.std(axis=1)[:, np.newaxis, :]
    X /= X.std(axis=1)[:, np.newaxis, :]
if division == 'std':
    TX /= X.std()
    X /= X.std()
```

```python
pylab.plot(X[:, Z.argmax(axis=1) == 0, 0])
print X.shape[1], TX.shape[1]
```

```python
n_hidden = [10]
hidden_transfer = ['tanh']
batch_size = X.shape[1]
batches_per_pass = X.shape[1] / batch_size
opt = 'rmsprop', {'steprate': 0.0001, 'decay': 0.8, 'momentum': 0.8, 'step_adapt': 0.01}
opt = 'rprop'
par_std = 1e-2
net = SupervisedRnn(X.shape[2], n_hidden, 2, hidden_transfer=hidden_transfer, 
                    optimizer=opt, batch_size=batch_size, pooling='mean',
                    loss='ncac')
net.parameters.data[...] = np.random.normal(0, par_std, net.parameters.data.shape)
bound_spectral_radius(net.parameters['recurrent_0'], 1.2)
```

```python
pause = modulo_n_iterations(10 * batches_per_pass)
stop = after_n_iterations(1000 * batches_per_pass)   

for info in net.powerfit((X, Z), (X, Z), stop, pause):
    print info['n_iter'], info['loss']#, info['step_rate'].mean()
```

```python
E, TE = net.predict(X), net.predict(TX)
```

```python
e, z = E, Z
colors = 'krmcrbg'
for i in range(Z.shape[1]):
    this_e = np.where(z.argmax(axis=1) == i)
    this_color = 'o' + colors[i]
    pylab.plot(e[this_e, 0], e[this_e, 1], this_color)
```
